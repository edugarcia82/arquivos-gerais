Uma análise para se obter algumas métricase insigths retirados de dois bancos de dados distintos, a proposta é seguir as premissas fornecidas pelo cliente, tratar os dados de acordo com essas premissas e extrair as informações requeridas após o tratamento.

```python
import pandas as pd
import seaborn as sns
```

Captura de tela 2023-06-24 230048.png

```python
#Carregando a tabela varejo para o programa.
vendas = pd.read_excel('/content/varejo.xlsx')
vendas.head(10)
```

```python
#Verificando as informações gerais da tabela, podemos ver que existem valores nulos nas colunas preço e estado, deverão ser tratado de acordo com as características do case
vendas.info()
```

```python
#Analisando as categorias de canais de venda, podemos constatar que existem categorias repetidas dentro da tabela (APP e Aplicativo).
vendas.groupby('idcanalvenda').idcompra.nunique()
```

```python
#Renomeando a categoria Aplicativo para APP e conferindo se a mudança foi efetuada.
vendas['idcanalvenda'] = vendas['idcanalvenda'].str.replace('Aplicativo', 'APP')
vendas.groupby('idcanalvenda').idcompra.nunique()
```

```python
#Verificando a coluna bandeira
vendas.groupby('bandeira').idcompra.nunique()
```

```python
#Verificando a coluna Nome_Departamento
vendas.groupby('Nome_Departamento').idcompra.nunique()
```

```python
#Verificando a coluna estado
vendas.groupby('estado').idcompra.nunique()
```

```python
#Substituindo os espaços dentro da coluna Nome_Departamento por underscore.
vendas['Nome_Departamento'] = vendas['Nome_Departamento'].str.replace(' ','_')
vendas.groupby('Nome_Departamento').idcompra.nunique()
```

```python
#testando a função dropna excluindo todas as linhas com valores nulos
vendas_sem_nulos = vendas.dropna()
vendas_sem_nulos.info()
```

```python
#Acrescentando a sigla correspondente ao estado do Mato Grosso do Sul nos campos nulos da coluna estado, conforme proposto no escopo do case.
vendas['estado'].fillna('MS', inplace = True)
vendas.info()
```

```python
#Substituindo os valores vazios da coluna Preço pela média de preço da tabela
media_preco = vendas['Preço'].mean()
vendas['Preço'].fillna(media_preco, inplace = True)
vendas.info()
```

```python
#verificando a premissa que dita a regra em que o Preço nunca pode ser maior que Preço com Frete
vendas[['Preço','Preço_com_frete']].describe()
```

```python
#declarando duas variáveis preco_errado que separa os preços certos e vendas_correto que eliina da tabela os preços que não atendem à premissa inicial de preço sempre menor que preço com frete.
preco_errado = vendas.query('Preço > Preço_com_frete')
vendas_correto = vendas.query('Preço < Preço_com_frete')
```

```python
#Nova contagem de linhas reduziu de 80228 para 22047.
vendas_correto.describe()
```

```python
vendas_correto.head()
```

```python
#Ordenando os departamentos com maiores números de vendas
vendas_correto.groupby('Nome_Departamento').idcompra.nunique().sort_values(ascending = False).reset_index()
```

```python
#Exibindo a média de preço com frete por departamento em ordem descrescente
round(vendas_correto.groupby('Nome_Departamento')['Preço_com_frete'].agg('mean').sort_values(ascending = False).reset_index(),2)
```

```python
#Exibindo a quantidade de vendas por data
vendas_correto.groupby('Data').idcompra.nunique().sort_values(ascending = False).reset_index()
```

```python
#Criando uma nova coluna com o mês da venda
vendas_correto['Mes'] = vendas_correto['Data'].dt.month
vendas_correto.head()
```

```python
#Exibindo a quantidade de vendas por mês.
vendas_correto.groupby('Mes').idcompra.nunique().sort_values(ascending = False).reset_index()
```

```python
#Carregando a tabela cliente_varejo para o programa.
cliente = pd.read_excel('cliente_varejo.xlsx')
cliente.head()
```

```python
#Comparando as colunas da tabela clientes_varejo e a tabela varejo
vendas_correto.head()
```

```python
round(cliente[['idade','renda']].describe(),1)
```

```python
#Verificando as informações contidas na tabela clientes.
cliente.info()
```

```python
#Transformando a coluna Renda em uma coluna do tipo float.
cliente = cliente.astype({'renda':'float'})
cliente.info()
```

```python
vendas_cliente = vendas_correto.merge(cliente, how = 'left', on = 'cliente_Log')
vendas_cliente.head()
```

```python
#Visualizando a renda média por categoria de acesso
agg_idcanal_renda = round(vendas_cliente.groupby("idcanalvenda")["renda"].agg('mean').sort_values(ascending=False).reset_index(),2)
agg_idcanal_renda
```

```python
#Media de idade dos clientes por bandeira
agg_idade_bandeira = round(vendas_cliente.groupby('bandeira')['idade'].agg('mean').sort_values(ascending=False).reset_index(),2)
agg_idade_bandeira.head()
```

```python
#Importando a biblioteca MatplotLIB
import matplotlib.pyplot as plt
```

```python
fig = plt.figure(figsize=(7,4))
plt.bar(agg_idade_bandeira['bandeira'], agg_idade_bandeira['idade'], color = 'green')
plt.ylabel('Média de Idade')
plt.title('Idade Média por Bandeira', loc = 'center', fontsize=18, fontweight='bold', color='black')
```

```python
fig = plt.figure(figsize=(7,4))
plt.bar(agg_idcanal_renda['idcanalvenda'], agg_idcanal_renda['renda'], color = 'green')
plt.ylabel('Renda Média')
plt.title('Renda Média por Canal de Venda', loc = 'center', fontsize=18, fontweight='bold', color='black')
```

```python
vendas_por_data = vendas_correto.groupby('Data').idcompra.nunique().reset_index()
vendas_por_data
```

```python
fig = plt.figure(figsize=(20,12))
plt.xlabel('Data da Venda')
plt.ylabel('Quantidade de Vendas')
x=vendas_por_data['Data']
y=vendas_por_data['idcompra']
plt.plot(x,y, color = 'green', linewidth = '2')
plt.show
```

```python
agg_dept_preco = round(vendas_correto.groupby('Nome_Departamento')['Preço_com_frete'].agg('mean').sort_values(ascending = False).reset_index(),2)
agg_dept_preco
```

```python
fig = plt.figure(figsize=(30,20))
plt.bar(agg_dept_preco['Nome_Departamento'], agg_dept_preco['Preço_com_frete'], color = 'green')
plt.ylabel('Preço Médio')
plt.title('Preço Médio por Departamento', loc = 'center', fontsize=18, fontweight='bold', color='black')
```
