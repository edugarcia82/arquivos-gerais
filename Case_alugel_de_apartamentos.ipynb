Esse programa busca análisar e entender um case onde o objetivo é encontrar uma opção de apartamento para locação que atenda as características e necessidades do locador.
Utilizaremos para isso, as bibliotecas Pandas e Seaborn, com o intuito de filtrar e tornar gráfico esse filtro.

```python
import pandas as pd
```

```python
aluguel_casa = pd.read_excel('/content/houses_rent.xlsx')
```

```python
aluguel_casa.head(10)
#verificando as 10 linhas iniciais
```

```python
aluguel_casa.tail(10)
#verificando as 10 linhas finais
```

```python
#renomeando colunas para retirar espaços
aluguel_casa = aluguel_casa.rename(columns={'parking spaces':'parking_spaces','rent amount':'rent_amount','property tax':'property_tax','fire insurance':'fire_insurance'})
```

```python
#verificando as 10 linhas iniciais
aluguel_casa.head(10)
```

```python
#verificando os tipos de dados em cada coluna
aluguel_casa.dtypes
```

```python
#modificando os tipos de dados nas colunas selecionadas
aluguel_casa = aluguel_casa.astype({'rent_amount':'float','property_tax':'float','fire_insurance':'float'})
aluguel_casa.dtypes
```

```python
#verificando as 10 linhas iniciais
aluguel_casa.head(10)
```

```python
#verificando as informações da tabela
aluguel_casa.info()
```

```python
#filtrando apartamentos com precificação de aluguel menor que 4000
aluguel_casa = aluguel_casa[(aluguel_casa['total'] <= 4000)]
aluguel_casa.head(10)
```

```python
#descrevendo as principais medidas da tabela
aluguel_casa.describe()
```

```python
#Adicionando filtros de permissão de animais
aluguel_casa = aluguel_casa[(aluguel_casa['total'] <= 4000) & (aluguel_casa['animal'] == 'accept')]
aluguel_casa.head()
```

```python
#utilizando a finção & e | para acrescentar filtros de número de quartos
aluguel_casa = aluguel_casa[(aluguel_casa['total'] <= 4000) & (aluguel_casa['animal'] == 'accept') & ((aluguel_casa['rooms'] == 2)|(aluguel_casa['rooms'] == 3)) ]
aluguel_casa.head()
```

```python
#descrevendo as principais contagens da tabela após as filtragens e comparando as diferenças com a descrição anterior
aluguel_casa.describe()
```

```python
#realizando as mesmas filtragens anteriores, porém, agora utilizando a função .query
aluguel_casa = aluguel_casa.query("total <= 4000 and animal == 'accept' and rooms == 2 or rooms ==3")
aluguel_casa.head()
```

```python
#agrupando os resultados por cidade
agg_city_price = aluguel_casa.groupby(['city'])['total'].agg('mean').reset_index()
agg_city_price
```

```python
#agrupando agora por número de banheiros e quartos
agg_city_price = aluguel_casa.groupby(['bathroom' , 'rooms'])['total'].agg('mean').reset_index()
agg_city_price
```

```python
#importando a biblioteca SEABORN para plotagem de gráficos para facilitar a visualização dos resultados
import seaborn as sns
```

```python
#plotando o gráfico de comparação entre os preços médios de alugueis de apartamentos de dois e 3 quartos
ax = sns.barplot(data=agg_city_price, x = 'rooms', y = 'total', color = 'blue');
ax.set(xlabel = 'qtd quartos', ylabel = 'média de preço');
```

```python
#plotando o gráfico de comparação entre os preços médios de alugueis de apartamentos de diferentes números de banheiro (HISTOGRAMA)
ax = sns.barplot(data=agg_city_price, x = 'bathroom', y = 'total', color = 'green');
ax.set(xlabel = 'qtd banheiros', ylabel = 'média de preço');
```
